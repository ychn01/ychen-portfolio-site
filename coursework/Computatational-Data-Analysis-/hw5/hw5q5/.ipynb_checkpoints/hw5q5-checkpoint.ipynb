```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV
from sklearn.linear_model import RidgeCV
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
```

```python
#load data and show image
mat = scipy.io.loadmat('cs.mat')
img = mat['img']

plt.imshow(img,  cmap='gray')
```

```python
#set A and y
np.random.seed(55)

A = np.random.randn(1300, 2500)
epsilon = np.random.normal(loc=0, scale=5, size=1300)
y = A.dot(img.ravel()) + epsilon
```

### Lasso Regression

```python
#lassoCV model with 10 fold cv
l_cv = LassoCV(cv=10, random_state=55).fit(A, y)
y_pred = l_cv.predict(A)
l_mse = mean_squared_error(y, y_pred)

#Plot Lasso cv curve
plt.figure(figsize=(8, 4))
plt.semilogx(l_cv.alphas_, l_cv.mse_path_.mean(axis=1), linewidth=2)
plt.xlabel('Alphas')
plt.ylabel('MSE')
plt.title('Lasso CV Error Curve')
plt.grid(True)
plt.show()
print('Optimal Lasso Alpha Value: ', l_cv.alpha_)
print('Lasso MSE: ', l_mse)

# recovered image
rec_img = l_cv.coef_.reshape(img.shape)
plt.subplot(1, 2, 2)
plt.imshow(rec_img, cmap='gray')
plt.title('Recovered Image using Lasso')
plt.tight_layout()
plt.show()
```

```python
alphas = np.arange(1, 201, 1)
```

```python
# Create Ridge Regression model
#alphas = np.logspace(-4, 4, 50)  # Range of alpha values to try
cv_mse = []

for alpha in alphas:
    ridge = Ridge(alpha=alpha)
    scores = -cross_val_score(ridge, A, y, cv=10, scoring='neg_mean_squared_error')
    cv_mse.append(np.mean(scores))

# Find optimal alpha and corresponding MSE
optimal_alpha = alphas[np.argmin(cv_mse)]
optimal_mse = min(cv_mse)

print("Best Alpha:", optimal_alpha)
print("Lowest MSE:", optimal_mse)

# Plot Cross-Validation Curve
plt.figure(figsize=(10, 6))
plt.plot(alphas, cv_mse, label='CV MSE')
plt.xlabel('Alpha')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Ridge Regression Cross-Validation (without Normalize)')
plt.axvline(optimal_alpha, color='r', linestyle='--', label='Optimal Alpha')
plt.legend()
plt.xscale('log')
plt.grid(True)
plt.show()

# Fit Ridge with the best alpha
ridge_model = Ridge(alpha=optimal_alpha)
ridge_model.fit(A, y)

# Get the coefficients and reshape to image size
recovered_image_ridge = ridge_model.coef_.reshape(50, 50)

# Plot the recovered image
plt.figure(figsize=(6, 6))
plt.imshow(recovered_image_ridge, cmap='gray')
plt.title('Recovered Image using Ridge Regression')
plt.axis('off')
plt.show()
```
