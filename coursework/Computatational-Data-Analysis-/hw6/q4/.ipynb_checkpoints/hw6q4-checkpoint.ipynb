```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
```

```python
mat = scipy.io.loadmat('data/data.mat')
data = mat['data']
X = data[:, 0]
y = data[:, 1] 
```

### b

```python
#function for local weighted linear reg
def linear_weighted_lr(x, X, y, tau, reg=1e-5):
    m = X.shape[0]
    weights = np.exp(-((X - x) ** 2) / (2 * tau ** 2))
    W = np.diag(weights)

    Xb = np.column_stack((np.ones(m), X))
    xb = np.append(1, x)

    bandwidth = np.linalg.inv(Xb.T @ W @ Xb + reg * np.identity(Xb.shape[1])) @ Xb.T @ W @ y

    return xb @ bandwidth
    
```

```python
def linear_weight_cv(X, y, bandwidths, n_folds=5):
    best_bandwidth = None
    best_score = float('inf')  # Initialize with a high value for minimization
    mse_values = []
    
    kf = KFold(n_splits=n_folds)
    
    for bandwidth in bandwidths:
        scores = []
        for train_index, val_index in kf.split(X):
            X_train, X_val = X[train_index], X[val_index]
            y_train, y_val = y[train_index], y[val_index]
            predictions = [linear_weighted_lr(x, X_train, y_train, bandwidth) for x in X_val]
            score = np.mean((predictions - y_val) ** 2)  # Mean squared error
            scores.append(score)
        avg_score = np.mean(scores)
        mse_values.append(avg_score)
        if avg_score < best_score:  # Lower MSE is better
            best_score = avg_score
            best_bandwidth = bandwidth
    
    return best_bandwidth, mse_values

# Define the range of bandwidth values to try
bandwidths = np.linspace(0.001, 2, 100)

# Generate some example data X and y
# You need to replace this with your actual dataset X and y

# Perform cross-validation to find the best bandwidth and obtain MSE values
best_bandwidth, mse_values = lwlr_cv(X, y, bandwidths)

print("Best bandwidth parameter (h):", best_bandwidth)

# Plot the cross-validation curve for MSE
plt.plot(bandwidths, mse_values, color='blue')
plt.xlabel('Bandwidth parameter (h)')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Cross-Validation Curve for Local Weighted Linear Regression')
plt.grid(True)

# Find the index of the minimum MSE value
min_mse_index = np.argmin(mse_values)

# Corresponding bandwidth value at the minimum MSE
best_bandwidth = bandwidths[min_mse_index]

# Minimum MSE value
min_mse = mse_values[min_mse_index]

# Plot marker at the lowest point
plt.scatter(best_bandwidth, min_mse, color='red', label=f'Best MSE: {min_mse:.4f} at h={best_bandwidth:.2f}')
plt.annotate(f'({best_bandwidth:.2f}, {min_mse:.4f})', (best_bandwidth, min_mse), textcoords="offset points", xytext=(0,10), ha='center')

plt.legend()
plt.show()
```
