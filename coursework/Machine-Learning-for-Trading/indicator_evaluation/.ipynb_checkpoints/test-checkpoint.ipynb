### JPM trades DF

```python
price = get_data([symbol], pd.date_range(sd, ed)).reset_index().rename(columns={'index': 'Date'})
del price['SPY']
price['Order'] = None
price.loc[0, 'Order'] = "BUY"
price['Hold'] = 0 
price.loc[0, 'Hold'] = 1000
price['Shares'] = 0
price.loc[0, 'Shares'] = 1000

for i, col in price[1:].iterrows():
    today_price = col[symbol]
    prev_hold = price.at[i-1, 'Hold']
    
    if i < len(price) - 1:
        tmrw_price = price.iloc[i + 1][symbol]
        if tmrw_price > today_price:
            if prev_hold == 1000:
                price.at[i, 'Order'] = "HOLD"
                price.at[i, 'Hold'] = prev_hold
                price.at[i, 'Shares'] = 0 
            elif prev_hold == 0: 
                price.at[i, 'Order'] = "BUY"
                price.at[i, 'Hold'] = prev_hold + 1000
                price.at[i, 'Shares'] = 1000 
            elif prev_hold == -1000:
                price.at[i, 'Order'] = "BUY"
                price.at[i, 'Hold'] = prev_hold + 2000
                price.at[i, 'Shares'] = 2000 
                
        elif tmrw_price < today_price:
            if prev_hold <= -1000:
                price.at[i, 'Order'] = "HOLD"
                price.at[i, 'Hold'] = prev_hold
                price.at[i, 'Shares'] = 0 
            elif prev_hold == 0:     
                price.at[i, 'Order'] = "SELL"
                price.at[i, 'Hold'] = prev_hold - 1000
                price.at[i, 'Shares'] = 1000 
            elif prev_hold == 1000:
                price.at[i, 'Order'] = "SELL"
                price.at[i, 'Hold'] = prev_hold - 2000
                price.at[i, 'Shares'] = 2000 
    else:
        price.at[i, 'Order'] = "HOLD"
        price.at[i, 'Hold'] = prev_hold
        price.at[i, 'Shares'] = 0 

trades_df = price.copy()
trades_df["Symbol"] = "JPM"
del trades_df["JPM"]
del trades_df["Hold"]

trades_df
# testpolicy = price[['Date', 'Shares']].copy()
# testpolicy.set_index('Date', inplace=True)
```

```python
import marketsimcode as msc
start_val = 100000
jpm_df = msc.compute_portvals(trades_df, start_val=100000, commission=0, impact=0)

jpm_df
```

### returns table

```python
benchmark = msc.compute_portvals(bm_df, start_val=100000, commission=0, impact=0)

def return_vals(port_val):
    dr = ((port_val/port_val.shift(1)) - 1)[1:]
    adr = round(dr.mean(),6)
    cr = round((port_val.iloc[-1]/port_val.iloc[0])-1,6)
    sddr = round(dr.std(), 6)
    df = pd.DataFrame({
        'Cumulative Return': cr,
        'Standard Deviation of Daily Return': sddr,
        'Average Daily Return': adr
    })
    df = df.reset_index(drop=True)
    df['Returns'] = ''
    return df

bench_vals = pd.DataFrame(return_vals(benchmark))
bench_vals['Returns'] = 'Benchmark'
port_vals = pd.DataFrame((return_vals(jpm_df)))
port_vals['Returns'] = 'Portfolio'

combined_df = pd.concat([bench_vals, port_vals])
combined_df.set_index('Returns', inplace=True)
print(combined_df)
display(combined_df)
```
