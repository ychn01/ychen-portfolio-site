```python
def compute_portvals(  		  	   		  		 		  		  		    	 		 		   		 		  
    orders_file="./orders/orders.csv",  		  	   		  		 		  		  		    	 		 		   		 		  
    start_val=1000000,  		  	   		  		 		  		  		    	 		 		   		 		  
    commission=9.95,  		  	   		  		 		  		  		    	 		 		   		 		  
    impact=0.005,  		  	   		  		 		  		  		    	 		 		   		 		  
):  		  	   		  		 		  		  		    	 		 		   		 		  
    """  		  	   		  		 		  		  		    	 		 		   		 		  
    Computes the portfolio values.  		  	   		  		 		  		  		    	 		 		   		 		  
  		  	   		  		 		  		  		    	 		 		   		 		  
    :param orders_file: Path of the order file or the file object  		  	   		  		 		  		  		    	 		 		   		 		  
    :type orders_file: str or file object  		  	   		  		 		  		  		    	 		 		   		 		  
    :param start_val: The starting value of the portfolio  		  	   		  		 		  		  		    	 		 		   		 		  
    :type start_val: int  		  	   		  		 		  		  		    	 		 		   		 		  
    :param commission: The fixed amount in dollars charged for each transaction (both entry and exit)  		  	   		  		 		  		  		    	 		 		   		 		  
    :type commission: float  		  	   		  		 		  		  		    	 		 		   		 		  
    :param impact: The amount the price moves against the trader compared to the historical data at each transaction  		  	   		  		 		  		  		    	 		 		   		 		  
    :type impact: float  		  	   		  		 		  		  		    	 		 		   		 		  
    :return: the result (portvals) as a single-column dataframe, containing the value of the portfolio for each trading day in the first column from start_date to end_date, inclusive.  		  	   		  		 		  		  		    	 		 		   		 		  
    :rtype: pandas.DataFrame  		  	   		  		 		  		  		    	 		 		   		 		  
    """  		  	   		  		 		  		  		    	 		 		   		 		  
    # this is the function the autograder will call to test your code  		  	   		  		 		  		  		    	 		 		   		 		  
    # NOTE: orders_file may be a string, or it may be a file object. Your  		  	   		  		 		  		  		    	 		 		   		 		  
    # code should work correctly with either input  		  	   		  		 		  		  		    	 		 		   		 		  
    # TODO: Your code here  		  	   		  		 		  		  		    	 		 		   		 		  
  		  	   		  		 		  		  		    	 		 		   		 		  
    # In the template, instead of computing the value of the portfolio, we just  		  	   		  		 		  		  		    	 		 		   		 		  
    # read in the value of IBM over 6 months  		  	   		  		 		  		  		    	 		 		   		 		  
    start_date = dt.datetime(2008, 1, 1)  		  	   		  		 		  		  		    	 		 		   		 		  
    end_date = dt.datetime(2008, 6, 1)  		  	   		  		 		  		  		    	 		 		   		 		  
    portvals = get_data(["IBM"], pd.date_range(start_date, end_date))  		  	   		  		 		  		  		    	 		 		   		 		  
    portvals = portvals[["IBM"]]  # remove SPY  		  	   		  		 		  		  		    	 		 		   		 		  
    rv = pd.DataFrame(index=portvals.index, data=portvals.values)  		  	   		  		 		  		  		    	 		 		   		 		  
  		  	   		  		 		  		  		    	 		 		   		 		  
    df = pd.read_csv(order_file)
    display(df)
    return rv  		  	   		  		 		  		  		    	 		 		   		 		  
    return portvals  		  	   		  		 		  		  		    	 		 		   		 		  
```

```python
df = pd.read_csv(orders_file)
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(by="Date", ascending = True)
syms = df['Symbol'].unique() 

start_date = np.datetime64(df['Date'].iloc[0])
end_date = np.datetime64(df['Date'].iloc[-1])
df = df.set_index('Date')

prices = get_data(syms, pd.date_range(start_date, end_date))



prices['Cash'] = 1 
print(prices)
daily_port = prices.copy() * 0 

print(daily_port)




holdings = prices * 0.0
holdings.iloc[0, -1] = start_val
for index, row in df.iterrows():
    stock = row[0]
    shares = row[2]
    price = prices.loc[index, stock]
    total_impact = shares * price * impact
    if row[1] == "BUY":
        buy_sell = 1
    else:
        buy_sell = -1
    share_amount = buy_sell * shares
    total_share_prices = share_amount * price
    holdings.loc[index, stock] += share_amount
    holdings.loc[index, "Cash"] -= total_share_prices
    holdings.loc[index, "Cash"] -= commission
    holdings.loc[index, "Cash"] -= total_impact


for date_row in range(1, holdings.shape[0]):
    for stock in range(0, holdings.shape[1]):
        new_stock_holding = holdings.iloc[date_row, stock] + holdings.iloc[date_row -1, stock]
        holdings.iloc[date_row, stock] = new_stock_holding

portfolio = prices * holdings

portfolio["port_val"] = portfolio.sum(axis = 1)
portfolio["daily_returns"] = (portfolio["port_val"][1:] /
portfolio["port_val"][:-1].values) - 1
portfolio["daily_returns"][0] = 0
portvals = portfolio.iloc[:, -2:-1]

display(portfolio)
display(portvals.head())
```
